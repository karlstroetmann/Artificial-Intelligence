# Minimization in PyTorch

```python
import torch
```

The function `f_and_fs` takes a floating point number `t` and computes the function
$$ f(t) := \exp(t) - 2 \cdot t^2 + 1 $$
It returns a pair $\bigl(f(\texttt{t}), f'(\texttt{t})\bigr)$, where $f'(\texttt{t})$ is the derivative of $\texttt{t}$.

```python
def f_and_fs(t):
    # declare x to be a tensor with value t that keeps track of its gradient
    x = torch.tensor(t, requires_grad=True, dtype=torch.double)
    y = torch.exp(x) - 2 * x * x + 1
    # use reverse mode automatic differentiation to compute the gradient of y w.r.t. x
    y.backward()
    # the gradient is stored in x.grad
    # the function item returns the value stored in a tensor
    return y.item(), x.grad.item()
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Create an array of t values
t_vals = np.linspace(-1, 3, 400)

# Compute f(t) using numpy
f_vals = np.exp(t_vals) - 2 * t_vals**2 + 1

# Create the plot
plt.figure(figsize=(8, 5))
plt.plot(t_vals, f_vals, label=r'$f(t) = \exp(t) - 2\cdot t^2 + 1$', color='blue')

# Formatting the plot
plt.title('Plot of f(t)')
plt.xlabel('t')
plt.ylabel('f(t)')
plt.axhline(0, color='black', linewidth=0.8)  # x-axis
plt.axvline(0, color='black', linewidth=0.8)  # y-axis
plt.grid(color='gray', linestyle='--', linewidth=0.5)
plt.legend()

# Display the plot
plt.show()
```

The function `findMinimum(start, eps)` computes the minimum of the function `f` defined above
using the method of gradient descent.
* `start` is the first value of the iteration.
* `eps` defines the precision.

```python
def findMinimum(start, eps):
    t     = start
    f, fs = f_and_fs(t)
    print(f'cnt = 0, f({t}) = {f}, fs = {fs}')
    alpha = 0.1   # learning rate
    cnt   = 0     # number of iterations
    while True:
        cnt += 1
        tOld, fOld = t, f
        t    -= alpha * fs
        f, fs = f_and_fs(t)
        print(f'cnt = {cnt}, f({t}) = {f}, fs = {fs}')
        if abs(t - tOld) <= abs(t) * eps:
            return t, f, fs, cnt            
        if f >= fOld:     # f didn't decrease, learning rate is too high
            alpha *= 0.5  # decrease the learning rate
            t, f = tOld, fOld    # reset t
            continue
        else:             # f has decreased
            alpha *= 1.2  # increase the learning rate
```

```python
findMinimum(1.0, 1e-12)
```
